1. 比較看看盒狀圖與 violin plot 在同樣的資料上分別展現資料的什麼特性。

2. 同樣的資料可以以條狀圖或是線圖繪製，思考兩者在表現資料上的特點。

```julia
# 模組import 

using RDatasets
using Gadfly
```

```julia
# retrieve data

iris = dataset("datasets", "iris")
longley = dataset("datasets", "longley")
```

```julia
# 1. boxplot
# 四分位距清楚, 但實際分布較為模糊，邊界點清楚。

plot(iris, x="Species", y="SepalWidth", Geom.boxplot)
```

```julia
# 1. violin
# 可以看得更清楚實際分布狀況, 但無法精確察覺四分位距

plot(iris, x = "Species", y = "SepalWidth", Geom.violin)
```

```julia
# 2. line
# 可以觀察資料的趨勢, 由線連結趨勢明顯
plot(longley, x = "Year", y="Employed", Geom.line)
```

```julia
# 2. dot
# 可以看出資料的真實分布狀況(對y)
plot(longley, x="Year", y="Employed", Geom.point)
```

```julia
# 2. bar

plot(longley, x="Year", y="Employed", Geom.bar)
```
